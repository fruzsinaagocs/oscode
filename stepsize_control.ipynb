```python
import numpy as np
import pyoscode
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
file1 = "examples/vdp_classic_control.txt"
file2 = "examples/vdp_PI_control.txt"

data1 = np.genfromtxt(file1, delimiter=',')
xs1 = data1[:, 0]
hs1 = data1[:, -2]
accepts1 = data1[:, -1]
data2 = np.genfromtxt(file2, delimiter=',')
xs2 = data2[:, 0]
hs2 =data2[:, -2]
accepts2 = data2[:, -1]
```

```python
plt.style.use('dense')
plt.figure(figsize=(6,4))
plt.semilogy(xs1[accepts1==1], hs1[accepts1==1], label='accepted', color='black', lw=1)
plt.semilogy(xs1[accepts1==0], hs1[accepts1==0], 'x', label='rejected', color='C1')
plt.legend()
plt.xlabel('$x$')
plt.ylabel('stepsize $h$')
plt.tight_layout()
plt.show()
#plt.savefig('thesiscorrs-vdp-classic.pdf')
```

```python
plt.style.use('dense')
plt.figure(figsize=(6,4))
plt.semilogy(xs2[accepts2==1], hs2[accepts2==1], label='accepted', color='black', lw=1)
plt.semilogy(xs2[accepts2==0], hs2[accepts2==0], 'x', label='rejected', color='C1')
plt.legend()
plt.xlabel('$x$')
plt.ylabel('stepsize $h$')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('dense')
fig, ax = plt.subplots(2, 1, figsize=(6,4), sharex=True)
ax[0].semilogy(xs1[accepts1==1], hs1[accepts1==1], label='accepted', color='black', lw=1)
ax[0].semilogy(xs1[accepts1==0], hs1[accepts1==0], 'x', label='rejected', color='C1')
ax[1].semilogy(xs2[accepts2==1], hs2[accepts2==1], color='black', lw=1)
ax[1].semilogy(xs2[accepts2==0], hs2[accepts2==0], 'x', color='C1')
ax[0].legend(loc='lower right')
ax[1].set_xlabel('$x$')
ax[0].set_ylabel('stepsize $h$')
ax[1].set_ylabel('stepsize $h$')
plt.tight_layout()
#plt.show()
plt.savefig('thesiscorrs-vdp-classic_and_PI.pdf')
```
